```python
# 欲しいプロットを作るために暗黙的かつ自動的に図形や軸を作成するインターフェース。
import matplotlib.pyplot as plt
import pandas as pd

# ワインデータ読み込み
wine = pd.read_csv("winequality-white.csv", sep =";", encoding = "utf-8")

# 品質データごとにグループ分けてカウント
# gropuby(グループ化する列)[計算対象の列]
count_data = wine.groupby('quality')["quality"].count()
# 一覧表
print(count_data)

# 数えたデータをグラフに描画
count_data.plot()
plt.savefig("wine-count-plg.png")
plt.show()
```
