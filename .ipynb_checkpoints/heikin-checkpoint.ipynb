```python
import pandas as pd

# Pandas でCSV読み込み
df = pd.read_csv("kion10y.csv", encoding ="utf-8")

# 日付ごとに気温をmdリストにまとめる
md = {}
# iterrows １行ずつ取得
for i, row in df.iterrows():

    m,d,v = (int(row['月']),int(row['日']),float(row['気温']))

    key = str(m) + "/" + str(d)

    # keyがないときは空の配列を代入
    if not(key in md): md[key] = []
    # 配列にv追加（気温）
    md[key] +=[v]

# 日付ごとに平均を求める
avs ={}
for key in md:
    v =avs[key] =sum(md[key]) / len(md[key])
    # '{0}, {1}, {2}...'.format(変数1, 変数2, 変数3….)  #インデックス（添え字）で指定
    print("{0}:{1}".format(key,v))
```
